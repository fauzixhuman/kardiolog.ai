## Butterworth Filter & Mel-Spectrogram
Membersihkan noise frekuensi rendah/tinggi dan memvisualisasikan energi suara.

```
def butter_bandpass(lowcut, highcut, fs, order=5):
    nyq = 0.5 * fs
    low = lowcut / nyq
    high = highcut / nyq
    b, a = butter(order, [low, high], btype='band')
    return b, a

# Simulasi proses
print('⚙️ Menjalankan Bandpass Filter (25-400Hz)...')
print('⚙️ Ekstraksi Mel-Spectrogram 2D...')
print('✅ Ekstraksi 33 Fitur Statistik (termasuk Shannon Energy) Selesai!')
```
